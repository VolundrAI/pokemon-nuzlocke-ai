```python
import requests
```

```python
def get_pokeapi_info(species, version_name="black-white"):
    """
    
    """
    
    pokedex_url = f"https://pokeapi.co/api/v2/pokemon/{species}/"
    pokespecies_url = f"https://pokeapi.co/api/v2/pokemon-species/{species}/"
    
    pokedex_entry = requests.get(pokedex_url).json()
    species_entry = requests.get(pokespecies_url).json()
    
    possible_abilities = [
        a["ability"]["name"]
        for a in pokedex_entry["abilities"]
        if not a["is_hidden"]
    ]
    
    odds_female = species_entry["gender_rate"] / 8
    
    moveset =  []
    for move in pokedex_entry["moves"]:
        for v_details in move["version_group_details"]:
            if v_details["move_learn_method"]["name"] == "level-up" and v_details["version_group"]["name"] == version_name:
                moveset.append((v_details["level_learned_at"], move["move"]["name"]))

    moveset = sorted(moveset, key=lambda x: x[0])
    
    return possible_abilities, odds_female, moveset

possible_abilities, odds_female, moveset = get_pokeapi_info("snivy")
```
